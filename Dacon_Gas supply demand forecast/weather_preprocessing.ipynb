```python

import numpy as np
import pandas as pd
import seaborn as sns
import os, glob, sys
import matplotlib.pyplot as plt

```

```python

from datetime import datetime

```

```python
d1 = pd.read_csv('./data/OBS_ASOS_TIM_20211017003024.csv',encoding='cp949')
d2 = pd.read_csv('./data/OBS_ASOS_TIM_20211017003024.csv',encoding='cp949')
d3 = pd.read_csv('./data/OBS_ASOS_TIM_20211017003024.csv',encoding='cp949')
d4 = pd.read_csv('./data/OBS_ASOS_TIM_20211017003024.csv',encoding='cp949')
d5 = pd.read_csv('./data/OBS_ASOS_TIM_20211017003024.csv',encoding='cp949')
d6 = pd.read_csv('./data/OBS_ASOS_TIM_20211017003024.csv',encoding='cp949')
d7 = pd.read_csv('./data/OBS_ASOS_TIM_20211017003024.csv',encoding='cp949')
```

```python
d = pd.concat([d1,d2])
d = pd.concat([d,d3])
d = pd.concat([d,d4])
d = pd.concat([d,d5])
d = pd.concat([d,d6])
d = pd.concat([d,d7])
```

```python
import datetime


display(d)
```

```python

d.rename(columns = {"일시": "time","기온(°C)" : "temperature", "습도(%)":"humidity", "증기압(hPa)" : "vapor", "풍속(m/s)":"wind_speed","현지기압(hPa)":"pressure"}, inplace = True)

d["time"] = d["time"].apply(lambda x : datetime.datetime.strptime(x,"%Y-%m-%d %H:%M" ))
d["temperature"] = d["temperature"].interpolate(method = "values")
d["humidity"] = d["humidity"].interpolate(method = "values")
d["vapor"] = d["vapor"].interpolate(method = "values")
d["wind_speed"] = d["wind_speed"].interpolate(method = "values")
d["pressure"] = d["pressure"].interpolate(method = "values")
```

```python
d = d[['time','temperature','humidity','vapor','pressure']]
```

```python
d.to_csv('./weather.csv')
```
